#  Exploration des données

# caracteristics

```python
df_caracteristics=pd.read_csv("../src/data/caracteristiques_2005a2021.csv",encoding="latin1",index_col=0)
df_caracteristics.head(5)
```

```python
df_caracteristics.info()
```

```python
df_caracteristics.describe()
```

```python
#chercher le % des valeurs NaN 
df_caracteristics.isna().sum() / len(df_caracteristics) * 100
```

```python
#on peut supprimer les colonnes des valeurs nan mais on garde la colonne atm (Conditions atmosphériques)

df_caracteristics = df_caracteristics.drop([ "adr", "gps", "lat", "long","dep","com"], axis=1)
```

```python
df_caracteristics.info()
```

```python
df_caracteristics['datetime_str'] = df_caracteristics['jour'].astype(str) + '/' + df_caracteristics['mois'].astype(str) + '/' + df_caracteristics['annee'].astype(str) + ' ' + df_caracteristics['hrmn'].astype(str)
df_caracteristics['datetime'] = pd.to_datetime(df_caracteristics['datetime_str'], format='%d/%m/%Y %H%M', errors='coerce')
df_caracteristics.drop(columns=['datetime_str'], inplace=True)
df_caracteristics
```

```python
df_caracteristics.info()
```

```python
# pour la colonne col (collision) on a va remplacer les valeurs manquantes
#on cherche les mod
set(df_caracteristics["col"].unique())
#Type de collision :
#-1 – Non renseigné
#1 – Deux véhicules - frontale
#2 – Deux véhicules – par l’arrière
#3 – Deux véhicules – par le coté
#4 – Trois véhicules et plus – en chaîne
#5 – Trois véhicules et plus - collisions multiples
#6 – Autre collision
#7 – Sans collision


#on peut remplacer les valeurs manquantes par 6:Autre collision

df_caracteristics["col"] = df_caracteristics["col"].replace(np.NaN, 6)
df_caracteristics.info()
```

```python
# pour la colonne atm (condition atmsphétique) on a va remplacer les valeurs manquantes
#on cherche les mod
set(df_caracteristics["atm"].unique())
#Type de collision :
#-1 – Non renseigné
#1 – Normale
#2 – Pluie légère
#3 – Pluie forte
#4 – Neige - grêle
#5 – Brouillard - fumée
#6 – Vent fort - tempête
#7 – Temps éblouissant
#8 – Temps couvert
#9 – Autre

#on peut remplacer les valeurs manquantes par 9:Autre 
df_caracteristics["atm"] = df_caracteristics["atm"].replace(np.NaN, 6)
df_caracteristics.info()
```
